```python
import matplotlib.pyplot as plt
d={'cs':82,'ai':75,'os':55,'se':62,'dda':77}
x=list(d.keys())
y=list(d.values())
fig=plt.figure(figsize=(10,5))
plt.bar(x,y,color=['blue','red','yellow','green','orange'])
plt.xlabel("courses of the semester")
plt.ylabel("obtained marks")
plt.title("mraks of semester of me")
plt.show()
```
